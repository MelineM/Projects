```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
import time
```

# Scraping & Cleaning

```python
url = "https://www.imdb.com/chart/top/?sort=rk,asc&mode=simple&page=1"
html = requests.get(url).content
soup = BeautifulSoup(html)
films = [i.get("href") for i in soup.select("a") if (i.get("href")!=None) and ("/title/t" in i.get("href"))]
links = ["http://www.imdb.com" + film for film in films]

title = []
iMDb_Rating = []
keywords = []

for link in links:
    time.sleep(0.2)
    url = link
    html = requests.get(url).content
    soup = BeautifulSoup(html)
    soup1 = [i.text for i in soup.select("h1")]
    title.append(soup1)
    soup2 = [i.text for i in soup.select("strong span")]
    iMDb_Rating.append(soup2)
    soup3 = [i.text for i in soup.select("a span.itemprop")]
    keywords.append(soup3)
    
Title = title[::2]
IMDb_Rating = iMDb_Rating[::2]
Keywords = keywords[::2]

Titles = []

for item in Title:
    for title in item:
        Titles.append(title.replace("(", " ").replace(")", "").replace("\xa0", ""))
        
year = []
Film_Title = []

for title in Titles:
    year.append(title[-5:])
    Film_Title.append(title[:-6])
    
Year= []

for item in year:
    Year.append(item.strip())

Rating = []

for rating in IMDb_Rating:
    for item in rating:
        Rating.append(item)   
```

```python
df = pd.DataFrame([Film_Title, Year, Rating, Keywords]).T
df.columns = ["Title", "Year", "IMDb_Rating", "Keywords"]
df
```

```python
#No missing data
df.isna().sum()
df.isnull().sum()
```

# Creating dummies for most common keywords

```python
keywordssss=list({j for i in Keywords for j in i})
low=[k for i in Keywords for j in i for k in j.split()]
dct={i:low.count(i) for i in set(low)}

revdct={k: v for k, v in sorted(dct.items(), key=lambda item: item[1], reverse=True)}
revdct
freq = {k:v for (k,v) in revdct.items() if v >= 5}

del freq["of"]
del freq["on"]
del freq["the"]
del freq["in"]
del freq["new"]
del freq["film"]
del freq["based"]
del freq["story"]
del freq["character"]
del freq["girl"]
del freq["woman"]
del freq["time"]
del freq["father"]
del freq["protagonist"]
del freq["one"]
del freq["registry"]
del freq["against"]
del freq["to"]
del freq["national"]
del freq["ending"]
del freq["a"]
del freq["male"]
del freq["year"]
del freq["man"]
del freq["false"]
del freq["as"]
del freq["two"]
del freq["person"]
del freq["post"]
del freq["from"]
del freq["city"]
del freq["human"]
del freq["female"]
del freq["child"]
del freq["son"]
del freq["sister"]

freq["nazi"] = freq["neo"] + freq["nazi"]
del freq["neo"]
freq["war"]  =  freq["battle"] + freq["war"] + freq["soldier"]
del freq["battle"]
del freq["soldier"]
freq["gang"]  =  freq["gangster"] + freq["gang"] + freq["mafia"]
del freq["gangster"]
del freq["mafia"]
freq["murder"]  =  freq["serial"] + freq["murder"] + freq["killer"]
del freq["serial"]
del freq["killer"]
freq["nudity"]  =  freq["frontal"] + freq["nudity"]
del freq["frontal"]
freq["british"]  =  freq["york"] + freq["british"]
del freq["york"]
freq["world"] = freq["mexico"] + freq["vietnam"] + freq["travel"]
del freq["mexico"]
del freq["vietnam"]
del freq["travel"]
freq
```

```python
df1 = df.copy()
for i in freq:
    df1[i]=df1.Keywords.map(lambda x: 1 if i in x else 0)
df1.drop("Keywords", axis=1, inplace=True)
df1.Year=df1.Year.astype(int)
df1.IMDb_Rating= df1.IMDb_Rating.astype(float)
df1.head(5)
```

```python
df1["Age^2"] = (2020 - df1.Year) ** 2
df1.drop("Year", axis=1, inplace=True)
df1.head(5)
```

# Supervised Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

### Linear Regression

```python
X_train,  X_test, y_train, y_test = train_test_split(df2,
                                                     df1.IMDb_Rating, test_size=0.33, random_state=42)
```

```python
linear_model = LinearRegression().fit(X_train, y_train)
y_test_pred = linear_model.predict(X_test)
y_test_pred
```

```python
A_vs_P = pd.DataFrame(columns=["Prediction", "Actual"])
A_vs_P['Prediction'] = y_test_pred
A_vs_P['Actual'] = list(y_test)

A_vs_P["Difference"] = A_vs_P['Prediction'] - A_vs_P['Actual']
round(A_vs_P, 2).head(10)
print("Average difference = ", round(A_vs_P["Difference"].mean(),2))
```

* Intercept = the expected mean value of Y when all X=0
* Coefficient/Slope = the % increase of rating for each change in a parameter

```python
print("Intercept = " + str(linear_model.intercept_))
coeffecients = pd.DataFrame(linear_model.coef_,X_test.columns)
coeffecients.columns = ["Coeffecient"]
coeffecients
```

1. R squared is a statistical measure of how close the data are to the fitted regression line. The closer to 1 the better.
2. Mean Absolute Error is the distance between data points and the line. The lower the better.
3. Mean Squared Error is the square of the distance between data points and the line. The lower the better.
4. Root Mean Squared Error is the square root of MSE. The lower the better.

```python
scoretest = r2_score(y_test, y_test_pred)
print("R-squared is ", scoretest)
mae = mean_absolute_error(y_test, y_test_pred)
print("MAE is:", mae)
mse = mean_squared_error(y_test, y_test_pred)
print("MSE is:", mse)
rmse = np.sqrt(mse)
print("RMSE is", rmse)
```

# Ensemble Algorithm

We have found the best algorithm for predicting IMDb user rating is the second Linear Regression algorithm but we can make strengthen the accuracy of our predictions by combining our algorithms in an ensemble algorithm.

Ensemble algorithms average out bias, reduce variance and are make overfitting unlikely. This produces better predictions!

```python
from sklearn.ensemble import VotingRegressor
```

```python
#create a dictionary of our models
estimators=[("k_model", k_model), ("rf_model", rf_model), ("tree_model", tree_model), 
            ("linear_model_2", linear_model_2), ("linear_model", linear_model)]
#create our voting classifier, inputting our models
ensemble = VotingRegressor(estimators)
```

```python
ensemble.fit(X_train4, y_train4)
```

```python
ensemble_r = (ensemble.score(X_test4, y_test4)*100)
ensemble_r
```

```python
df7 = pd.DataFrame({
    "Model": ["Linear Regression1", "Linear Regression2", "Decision Tree", "Random Forest","K Nearest Neighbour",
              "Average", "Ensemble"],
    "R2": [scoretest*100, scoretest1*100, scoretest2*100, scoretest3*100, scoretest4*100,
           ((scoretest+scoretest1+scoretest2+scoretest3+scoretest4)/5)*100, ensemble_r,],
})
round(df7, 2)
```
